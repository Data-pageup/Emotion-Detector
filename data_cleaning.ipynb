```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'E:\PG-projects\6_churm prediction\customer_churn_dataset.csv')
```

# Removing unwanted columns

```python
df.drop(columns = ['customer_id'], inplace = True)
```

```python
df.isnull().sum()
```

# Encoding 

```python
df.contract.unique()
```

```python
mapping = {'Month-to-month': 0,
    'One year': 1,
    'Two year': 2}
df['contract'] = df['contract'].map(mapping)
```

```python
df.payment_method.unique()
```

```python
df = pd.get_dummies(df, columns=['payment_method'], prefix ='pay')
```

```python
df.columns
```

```python
df.internet_service.unique()
```

```python
df['internet_service'] = df['internet_service'].fillna('No')
df = pd.get_dummies(df, columns=['internet_service'], prefix = 'internet_service')
```

```python
df.columns
```

```python
df.tech_support.unique()
```

```python
mapping = {'No': 0, "Yes":1}

df['tech_support'] = df['tech_support'].map(mapping)
```

```python
df.online_security.unique()
```

```python
df['online_security'] = df['online_security'].map(mapping)
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
df.to_csv('cleaned_tele_churn.csv', index = False)
```
